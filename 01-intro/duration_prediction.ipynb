```python
df = pd.read_parquet('/workspaces/mlops_zoomcamp/data/green_tripdata_2021-01.parquet')
```

```python
df.head()
```

```python
# parse to datetime
df.lpep_dropoff_datetime = pd.to_datetime(df.lpep_dropoff_datetime)
df.lpep_pickup_datetime = pd.to_datetime(df.lpep_pickup_datetime)
```

```python
# duration
df["duration"] = df.lpep_dropoff_datetime - df.lpep_pickup_datetime
df["duration"] = df.duration.apply(lambda td: td.total_seconds()/60) # Have the results in minutes
```

```python
df.head()
```

```python
# filter data
df = df[((df.duration > 1) & (df.duration <= 60))]
```
